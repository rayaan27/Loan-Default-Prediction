```python
import joblib
import os

files_to_compress = [
    'model_a.pkl',
    'model_b.pkl',
    'best_model.pkl',
    'shap_explainer.pkl'
]

results = []

for filename in files_to_compress:
    path = f'../models/{filename}'
    
    if not os.path.exists(path):
        print(f"{filename}: NOT FOUND, skipping")
        continue
    
    size_before = os.path.getsize(path) / (1024*1024)
    
    obj = joblib.load(path)
    joblib.dump(obj, path, compress=3)
    
    size_after = os.path.getsize(path) / (1024*1024)
    
    reduction_pct = (1 - size_after/size_before) * 100
    results.append((filename, size_before, size_after, reduction_pct))
    print(f"{filename}: {size_before:.2f} MB -> {size_after:.2f} MB  ({reduction_pct:.1f}% reduction)")

print("\n--- Summary ---")
for filename, before, after, pct in results:
    status = "✅ Under 100MB" if after < 100 else "❌ Still over 100MB"
    print(f"{filename}: {after:.2f} MB — {status}")
```

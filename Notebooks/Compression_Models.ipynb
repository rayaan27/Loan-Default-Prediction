```python
import joblib
import os

files_to_compress = ['model_a.pkl', 'model_b.pkl']

for filename in files_to_compress:
    path = f'models/{filename}'
    
    # Check size before
    size_before = os.path.getsize(path) / (1024*1024)
    
    # Load and re-save with compression
    model = joblib.load(path)
    joblib.dump(model, path, compress=3)
    
    # Check size after
    size_after = os.path.getsize(path) / (1024*1024)
    
    print(f"{filename}: {size_before:.2f} MB -> {size_after:.2f} MB")
```
